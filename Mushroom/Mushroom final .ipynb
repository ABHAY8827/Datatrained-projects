```python
from sklearn.feature_selection import SelectPercentile
from sklearn.feature_selection import chi2
```

```python
#Instantiate selectpercentile and fit(feature,label)
X=df.drop(columns=['class'],axis=1)
y=df['class']

Spercentile = SelectPercentile(score_func=chi2, percentile=80)
Spercentile.fit(X,y)
```

```python
# seperate the features to check p_values
cols = Spercentile.get_support(indices=True)
features = X.columns[cols]
```

```python
df_scores = pd.DataFrame({'features':X.columns, 'Chi2score':Spercentile.scores_, 'pvalue':Spercentile.pvalues_,})
df_scores.sort_values(by='Chi2score',ascending=True)
```

```python
print(cols)
print(features) #this are top 80% features
```

```python
X = df[features]
y = df['class']
```

```python
import warnings
warnings.filterwarnings('ignore')
```

# model building

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
maxAcc = 0 #maximum accuracy
maxRs = 0 #best random state value for which max accuracy is achieved 
for i in range(1,200):
    x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=.20,random_state=i)
    LR = LogisticRegression()
    LR.fit(x_train,y_train) #fitting the data will train your model
    predlr = LR.predict(x_test) #this is the predicted target variable
    acc = accuracy_score(y_test,predlr)
    if acc>maxAcc:
        maxAcc = acc
        maxRs = i
print("Best accuracy is",maxAcc,"on random state",maxRs)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=180)
```

### Logistic Regression

```python
from sklearn.metrics import classification_report,roc_auc_score,roc_curve
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(x_train,y_train)
predlr = LR.predict(x_test)
acc1 = accuracy_score(y_test,predlr)
print(acc1)
```

```python
print(classification_report(y_test,predlr))
```

### Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
pred_dt = dt.predict(x_test)
acc2 = accuracy_score(y_test,pred_dt)
print(acc2)
print(classification_report(y_test,pred_dt))
```

### Random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
pred_rf = rf.predict(x_test)
acc3 = accuracy_score(y_test,pred_rf)
print(acc3)
print(classification_report(y_test,pred_rf))
```

### Gradient Boosting classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier()
gb.fit(x_train,y_train)
pred_gb = gb.predict(x_test)
acc4 = accuracy_score(y_test,pred_rf)
print(acc4)
print(classification_report(y_test,pred_gb))
```

```python
from xgboost import XGBClassifier
xgbt = XGBClassifier()
xgbt.fit(x_train,y_train)
pred_xgbt = xgbt.predict(x_test)
acc5 = accuracy_score(y_test,pred_rf)
print(acc5)
print(classification_report(y_test,pred_xgbt))
```

## Roc-Auc curve

```python
from sklearn.metrics import plot_roc_curve
disp = plot_roc_curve(dt, x_test, y_test)
plot_roc_curve(LR, x_test, y_test, ax = disp.ax_)
plot_roc_curve(rf, x_test, y_test, ax = disp.ax_)
plot_roc_curve(gb, x_test, y_test, ax = disp.ax_)
plot_roc_curve(xgbt, x_test, y_test, ax = disp.ax_)


plt.show()
```

```python
# as we are getting very high accuracy we will check it with cross validation for model overfitting
```

## Cross validation

```python
from sklearn.model_selection import cross_val_score
scr = cross_val_score(LR, X, y, cv=5)
print("Cross validation score of logistic regression model is",scr.mean())
```

```python
from sklearn.model_selection import cross_val_score
scr = cross_val_score(dt, X, y, cv=5)
print("Cross validation score of Decision tree model is",scr.mean())
```

```python
from sklearn.model_selection import cross_val_score
scr = cross_val_score(gb, X, y, cv=5)
print("Cross validation score of Gradient boosting model is",scr.mean())
```

```python
from sklearn.model_selection import cross_val_score
scr = cross_val_score(xgbt, X, y, cv=5)
print("Cross validation score of xtreme Gradient boosting model is",scr.mean())
```

## from this we can see best model is Gradient boosting model soo we will take it ahead and perform hyperparameter tuning

### GridSearchCv

```python
from sklearn.model_selection import GridSearchCV

#creating parameter list to pass in Gridsearchcv
parameter ={'learning_rate': np.arange(0.1,0.5,0.1),
           'max_depth':range(4,8),
            'min_samples_split':range(2,6)
           }
```

```python

GCV = GridSearchCV(gb,parameter,cv=5,scoring='accuracy')

GCV.fit(x_train,y_train)  #fitting the data in model
GCV.best_params_          #printing the best parameter found by Gridsearchcv
GCV_pred = GCV.best_estimator_.predict(x_test)    #predicting with best parameter 
accuracy_score(y_test, GCV_pred)                  #checking final accuracy
```
